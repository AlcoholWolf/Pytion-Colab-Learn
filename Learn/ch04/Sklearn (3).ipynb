````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 15
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['image.cmap'] = 'rainbow'
plt.style.use('dark_background')

def XS(min, max, lim): # ys
    xs = np.linspace(min,max,lim)
    return xs

def OSK(x): # f(x)=x^2+2x+1
    fx = (x**2)+(2*x)+1
    return fx

def COS(x): # f(x) = cosin(x)
    fx = np.cos(x)
    return fx

def MSK(x, y): # f(x) = x + y
    fx = x + y
    return fx

def MSV(x, y): # f(x) = (x-5)^2 + (y-5)^2
    fx = (x-5)**2 + (y-5)**2
    return fx

def MKV(x, y): # fx = 50(y-x^2)^2+(2-x)^2
    fx = 50*(y-(x**2))**2+(2-x)**2
    return fx

def OKV(x): # 일변수 벡터
    fx1, fx2 = np.array([x**2 , 5*x])
    return fx1, fx2

def MDV(x): # 다변수 벡터
    fx = np.array([np.sin(6*x), 1/4*x, x**2/2])
    return fx

def CST(x):
    return np.array([np.cos(x), np.sin(x), x])

def MPV(x, y):
    return np.array([x**2, y**2, x*y])

def MUV(u, v):
    return np.array([u, v, 1+u**2+v/(1+v**2)])



def MSVEAU2D(x, s): # x : x값 / s : 함수
    fig, ax = plt.subplots(1)
    ax.plot(x, s(x))
    
    ax.set_title(f'def 2D')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    # x = -1
    # ax.plot(x, s(x), marker='o', c='r')
    # ax.text(x, s(x), f'({x} . {np.round(s(x),2)})')

    return plt.show()

def MSVEAU2DV(x, s): # x : x값 / s : 함수
    fig, ax = plt.subplots(1)

    ax.plot(s(x)[1], s(x)[0])
    ax.set_title('$f(x)$ space')
    ax.set_xlabel('$g(x)$')
    ax.set_ylabel('$h(x)$')

    scat = ax.scatter(s(x)[1],s(x)[0], c=x, cmap='rainbow')
    fig.colorbar(scat)

    return plt.show()

def MSVEAU3D(xs, ys, s): # x : x값 / y : y값 / s : 함수
    x, y = np.meshgrid(xs, ys)

    fig = plt.figure(figsize=(9, 7))
    ax = plt.axes(projection='3d')
    ax.scatter3D(x, y, s(x, y), c='white', s=60)
    CF = ax.plot_surface(x, y, s(x, y), cmap='rainbow')
    
    ax.set_title(f'def 3D')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    # x = -1
    # ax.plot(x, s(x), marker='o', c='r')
    # ax.text(x, s(x), f'({x} . {np.round(s(x),2)})')
    
    # fig.colorbar(CF, label='CMAP', shrink=0.8)

    return plt.show()

def MSVEAU3DV(ts, s):
    xs, ys, zs = s(ts)
    fig = plt.figure(figsize=(10,7))
    ax = plt.axes(projection='3d')
    scat = ax.scatter3D(xs, ys, zs, c=ts, marker='o', cmap='rainbow')

    ax.set_title('$f(t)$ space')
    ax.set_xlabel('$x(t)$')
    ax.set_ylabel('$y(t)$')
    ax.set_zlabel('$z(t)$')
    fig.colorbar(scat, shrink=.6, label='ts')
    return plt.show()

def MSVEAU3DD(ts, s):
    tts = ts
    X, Y = np.meshgrid(tts, ts)
    Fs = s(X, Y)
    fig = plt.figure(figsize=(10,7))
    ax = plt.axes(projection='3d')
    ax.scatter3D(Fs[0], Fs[1], Fs[2], marker='.')
    return plt.show()
````

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 2d플롯으로 시각화

````python
MSVEAU2DV(XS(-10, 10, 20), OKV)
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 3d플롯으로 시각화

````python
MSVEAU3DV(XS(-1, 3, 200), MDV)
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* 3d플롯으로 시각화

````python
MSVEAU3DV(XS(-10, 10, 200), CST)
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 
  * cos(t)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * sing(t)가 가질 수 있는 값 (두번째 아웃풋) : 


## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 3d플롯으로 시각화

````python
MSVEAU3DD(XS(-0.5,1,50), MPV)
````

\end{align}

* 3d플롯으로 시각화
