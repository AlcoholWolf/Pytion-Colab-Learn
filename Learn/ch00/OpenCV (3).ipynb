### **| 색깔 파라미터 설정**

````python
# G, B, R  : 0 ~ 255
R = 250 #@param {type:"slider", min:0, max:255, step:10}
G = 250 #@param {type:"slider", min:0, max:255, step:10}
B = 250 #@param {type:"slider", min:0, max:255, step:10}
shape = (100,100,3)
COLOR = (B, G, R)
bkg = np.full(shape, COLOR, np.uint8) # 백그라운드 
cv2_imshow(bkg)
````

## 도형 그리기
---

### **| Line 그리기**

Start와 End 점을 연결하여 직선을 그립니다.

#### `cv2.line(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지 파일
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

````python
cv2.line(img, (X, Y), (X+SIZE, Y+SIZE), COLOR, 10)
cv2_imshow(img)
````

### **| 사각형 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.rectangle(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

* Sample Code
```
img = cv2.rectangle(img, (384, 0), (510, 128), (0,255,0), 3)
```

````python
cv2.rectangle(img, (X, Y), (X+SIZE, Y+SIZE), COLOR, 5)
cv2_imshow(img)
````

### **| 원 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.circle(img, center, radian, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * center – 원의 중심 좌표(x, y)
  * radian – 반지름
  * color – BGR형태의 Color
  * thickness – 선의 두께, -1 이면 원 안쪽을 채움

* Sample Code
```
img = cv2.circle(img, (447,63), 63, (0,0,255), -1)
```

````python
cv2.circle(img, (X, Y), 50, COLOR, 5)
cv2_imshow(img)
````

### **| 텍스트 추가**

#### `cv2.putText(img, text, org, font, fontSacle, color, thickness)`
* Parameters
 * img – image
 * text – 표시할 문자열
 * org – 문자열이 표시될 위치. 문자열의 bottom-left corner점
 * font – font type. CV2.FONT_XXX
 * fontSacle – Font Size
 * color – fond color
 * thickness - 선의 두께

* Sample Code
```
cv2.putText(img, 'OpenCV', (10,500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255,255,255), 2)
```

````python
cv2.putText(img, 'OpenCV', (X, Y), cv2.FONT_ITALIC, 2, COLOR, 4)
cv2_imshow(img)
````

#### **예제 01**

````python
shape = (200, 200, 3)
value = 255
bkg = np.full((shape), value, np.uint8)
plt.imshow(bkg)

SIZe = 20

cv2.line(bkg, (40, 50), (55, 70), (20, 200, 100), 2)
cv2.putText(bkg, 'Hello', (90, 70), cv2.FONT_ITALIC, 1, (20, 200, 100), 1)
cv2.rectangle(bkg, (10, 50), (10+SIZe, 50+SIZe), (20, 200, 100), 2)
cv2.circle(bkg, (70, 60), 10, (20, 200, 100), 2)
plt.imshow(bkg)
````

````python
shape = (100, 100, 3)
value = 255
bkg = np.full((shape), value, np.uint8)
plt.imshow(bkg)

x = 0
y = 0
s = 20

for I in range(5):
    cv2.rectangle(bkg, (x, y), (x+s, y+s), (20, 200, 100), 2)
    x += 20
    y += 20

cv2.circle(bkg, (80, 20), 10, (20, 200, 100), 20)
cv2.circle(bkg, (20, 80), 10, (20, 200, 100), 20)

plt.imshow(bkg)
````
