#### **랜덤값 생성**



* `random.random` : 0과 1사이의 난수를 무작위로 생성합니다.

```
random.random()
```



````python
A = 0
while True:
  A = random.random()
  print(A)
  if A == 1:
    break
````

* `random.randint`: 특정 범위 랜덤한 정수를 무작위로 생성합니다. (a이상 b 이하)

```
random.randint(a, b)
```



````python
random.randint(1, 6)
````

#### **랜덤값 추출**

* `random.choice`: 주어진 sequence 에서 랜덤으로 한개의 요소를 뽑습니다.

```
a = [1,2,3] # 리스트 a가 주어졌을 때 
b = random.choice(a) # a의 요소 중 한개를 뽑아 b로 바인딩
print(b)
```



````python
a = [1,2,3]
b = random.choice(a)
print(b)
````

* `random.choices` : 주어진 sequence 에서 랜덤으로 k개의 요소를 뽑습니다. 중복이 가능합니다.

```
a = [1,2,3] # 리스트 a가 주어졌을 때 
b = random.choices(a, k=2) # a의 요소 중 k개를 뽑아 b로 바인딩
print(b)
```



````python
a = [1,2,3]
b = random.choices(a, k=2)
print(b)
````

#### **기타**

* `random.shuffle` : 특정 시퀀스를 랜덤하게 섞는 함수입니다.

```
a = [1,2,3]
random.shuffle(a)
print(a)
```



````python
a = [1,2,3]
random.shuffle(a)
print(a)
````

####**예제 01 여러번의 주사위**


    - 주사위를 던지면 1에서 6까지 숫자가 랜덤으로 나옵니다.
    - 주사위를 사용하가 입력한 만큼 돌리고 각숫자가 몇번씩 나왔는지 알려주는 함수를 작성하세요


````python
A = int(input())
B = []
while True:
  A -= 1
  B.append(random.randint(1, 6))
  if A == 0:
    break
C1 = B.count(1)
C2 = B.count(2)
C3 = B.count(3)
C4 = B.count(4)
C5 = B.count(5)
C6 = B.count(6)
print(f' * {B} \n * 1: {C1}번  2: {C2}번  3: {C3}번  4: {C4}번  5: {C5}번  6: {C6}번')
````

####**예제 02 업다운게임**

    - 1~100까지 값이 랜덤하게 지정됩니다.
    - 랜덤값(정답)이 사용자가 찍은 답보다 높으면 UP
    - 랜덤값이 낮으면 DOWN
    - 랜덤값을 맞추면 "CORRECT"를 출력합니다.
    - 사용자가 랜덤값을 맞출 때까지 진행됩니다.


````python
print('게임 목록 : UDS')
A = input('게임 선택하기 : ')
if A =='UDS':
  print('램덤한 숫자가 적용되엇습니다.')
  B = random.randint(1, 100)
  while True:
    C = int(input('숫자를 입력하십시오. : '))
    if C == B:
      print('정답입니다!')
      break
    elif C < B:
      print('Up!')
    elif C > B:
      print('Dawn!')
    elif C >= 101 or C <= 0:
      print('1~100 사이의 값을 골라주십시오.')
````

####**예제 04 같은 그림 맞추기**
