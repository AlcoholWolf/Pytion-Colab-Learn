### **01. 플롯 분할**
---

````python
# df.plot(x='M',y='C')
int1 = 1
x = np.linspace(-5, 5, 5000)
def afu1(a):
  y = a*(x**2)
  return y
def afu2(a):
  y = -a*(x**2)
  return y
def afu3(a):
  y = (-2*a)*(x**2)
  return y
def afu4(a):
  y = (2*a)*(x**2)
  return y

plt.subplot(1,2,1)
plt.plot(x, afu1(int1))
plt.plot(x, afu4(int1))
plt.subplot(1,2,2)
plt.plot(x, afu2(int1))
plt.plot(x, afu3(int1))
plt.show()
````

### **02. 눈금 공유**
---



````python
t = np.arange(0.01, 5, 0.01)
s1 = np.sin(t*np.pi)
s2 = np.sin(t**np.pi)
s3 = np.exp(-t)

ax1 = plt.subplot(3,1,1)
plt.plot(t, s1)
plt.tick_params('x', labelbottom=False)

plt.subplot(3,1,2, sharex=ax1)
plt.plot(t, s2)
plt.tick_params('x', labelbottom=False)

plt.subplot(3,1,3, sharex=ax1)
plt.plot(t, s3)

plt.show()
````

### **01. 로그함수와 지수함수 그리기**
---

````python
t = np.arange(0.01, 5, 0.01)
x = np.linspace(-5, 5, 500)

l1 = np.log(t)
l2 = np.exp(t)

as1 = plt.subplot(2, 2, 4)
plt.plot(t, l1)
plt.tick_params('y', labelleft=False)

plt.subplot(2, 2, 1, sharex=as1)
plt.plot(t, l2)
plt.tick_params('x', labelbottom=False)

as2 = plt.subplot(2, 2, 3)
plt.plot(-t, l1)

plt.subplot(2, 2, 2, sharex=as2)
plt.plot(-t, l2)
plt.tick_params('x', labelbottom=False)
plt.tick_params('y', labelleft=False)

plt.show()
````

````python
df = sns.load_dataset('flights')
# plt.plot((df['month']), df['passengers'], linewidth='1')
plt.bar(df['month'], df['passengers'], width=0.6)
plt.bar(df['month'], df['passengers'], width=0.4)
````

````python
df.sort_values("year",ascending=False,inplace=True)
p2 = df.pivot_table(index="month", columns="year", values="passengers", sort=True)
p2.plot(figsize=(15,7))
plt.ylabel("passengers", labelpad=7, fontsize=20)
plt.xlabel("month", labelpad=7, fontsize=20)
plt.legend(title = "years",bbox_to_anchor=(0.58, 0, 0.5, 0.5))
plt.show()
````
