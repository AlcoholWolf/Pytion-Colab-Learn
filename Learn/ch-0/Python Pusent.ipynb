```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]                 # 출력 크기 (x12 y8)
plt.style.use('dark_background')                         # 메타포트 레이아웃 - 검은 바탕
```

#### **예제 01. 공정한 주사위**



* 확률변수가 '주사위를 굴려 나온 눈의 값' 일 때, 확률변수 $X$를 정의하세요.

```python
X = [1,2,3,4,5,6]
```

* 주사위의 모든 면이 '공정' 할 때, 확률 질량 함수 $P$를 정의하세요.

```python
def P(x):
    if 1 <= x <= 6:
        return 1/6
    else:
        pass

P_x = list(map(P, X))
print(P_x)
```

* 확률 분포를 그리세요.

```python
_=plt.bar(X, P_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 02. 동전 두번 던지기**


* 확률변수가 '앞면이 나온 횟수' 일 때, 확률변수 $X$를 정의하세요.

```python
X = [0,1,2]
```

* 확률 질량 함수를 정의하세요.

```python
def P(x):
  if x==0: p=.25
  elif x==1: p=.5
  elif x==2: p=.25
  else: p=0
  return p

P_x = list(map(P, X))
print(P_x)
```

* 확률 분포를 그리세요.

```python
_=plt.bar(X, P_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 03. 화재 발생**


*  확률변수가 '1년간 발생한 화재건수' 일 때 $X$를 정의하세요.

```python
X = np.arange(20)
print(X)
```

* 화재발생 확률이 다음과 같이 주어졌을 때 함수 $f$ 를 정의하세요.

\begin{align}
f(X=x)=\frac{1}{2^{(x+1)}}
\end{align}

```python
def f(x):
  return 1/(2**(x+1))

f_x = f(X)
print(f_x)
```

* 확률 분포를 그리세요.

```python
_=plt.bar(X, f_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(f_x), marker='o')
```

#### **예제 04. 공정하지 않은 주사위**


* 확률변수가 '주사위를 굴려 나온 눈의 값' 일 때, 확률변수 $X$를 정의하세요.

```python
X = np.arange(1,7)
print(X)
```

* 주사위의 짝수 면이 나올 확률이 홀수 면이 나올 확률의 2배일 때, 확률 질량 함수 $P$를 정의하세요.

```python
def P(x):
    if (x%2) == 1:
        return 1/9
    elif (x%2) == 0:
        return 2/9
    else:
        pass

P_x = list(map(P, X))
print(P_x)
```

* 확률분포 를 그리세요.

```python
_=plt.bar(X, P_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 05. 주사위 두번 굴리기**


* 확률변수가 '주사위를 두번 굴려 나온 눈의 합' 일 때, 확률변수 $X$를 정의하세요.

```python
Sp = [1,2,3,4,5,6]
```

*  확률 질량 함수 $P$ 를 정의하세요.

```python
Pp = []
Ps = []
PPS = []

for sum_s1 in Sp:
    for sum_s2 in Sp:
        Pp.append(sum_s1+sum_s2)
        Ps.append((sum_s1, sum_s2))
Pps = list(set(Pp))
for ii in Pps:
    PPS.append(Pp.count(ii))

print(Pps)
print(PPS)
```

* 확률 분포를 그리세요.

```python
_=plt.bar(Pps, PPS)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(PPS), marker='o')
```

#### **예제 06. 손의 길이**


* 확률변수가 '성인의 손 길이(cm)' 입니다. $X$를 10부터 20까지 1000개의 값으로 정의하세요.

```python
X = np.linspace(10,20,1000)
```

* 손 길이에 대한 확률이 다음과 같이 주어졌을 때 함수 $f$ 를 정의하세요. ($\sigma$ = 1)



\begin{align}
 f(x)=\frac{1}{\sqrt{2\pi}\sigma}\exp^{-\frac{(x-15.2)^2}{2*\sigma^2}}
\end{align}

```python
def f(x, sigma=1):
    return ( 1/np.sqrt(2*np.pi)*sigma ) * np.exp( -  (x-15.2)**2 / (sigma)**2 )
```

* 확률 분포를 그리세요.

```python
P_x = f(X)
_ = plt.plot(X, P_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(P_x))
```

#### **예제 07. 파라미터 변경**

예제 06 에서 정의한 함수 $f$의 파라미터 $\sigma$를 각각 `0.5, 1, 2, 4`로 정의한 후 확률분포와 누적확률 분포를 그리세요.

\begin{align}
 f(x)=\frac{1}{\sqrt{2\pi}\sigma}\exp^{-\frac{(x-15.2)^2}{2*\sigma^2}}
\end{align}

```python
fig, axes = plt.subplots(2,4, figsize=(15,6))
axes = axes.flatten()
for i, sigma in enumerate([0.5,1,2,4]):
  axes[i].plot(X, f(X,sigma=sigma))
  axes[i].set_title(f'sigma : {sigma}')  
  axes[i+4].plot(np.cumsum(f(X,sigma=sigma)))
```
